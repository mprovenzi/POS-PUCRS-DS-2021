Exemplo básico utilizando um modelo de Regressão Logística para prever se, dada uma determinada temperatura externa, uma determinada pessoa deve ou não levar casaco. 

```python
from sklearn.linear_model import LogisticRegression   # Importando modelo de Regressão Logística do SciKit Learn
```

```python
# Dados para treinamento: tabelas verdade preenchidas pelo usuário a partir da sua experiência
temperature =   [[30],  [12], [14], [18], [25],   [5], [15], [27]]  # Temperaturas em oC
coat =          [False, True, True, True, False, True, True, False] # True = Leva casaco; 
```

```python
reg1 = LogisticRegression() # cria o modelo de Reg Log
reg1.fit(temperature, coat) # treina a ML
```

```python
# Verifica se leva ou não casaco em uma série de temperaturas diferentes das de treino
reg1.predict([[10], [1], [20], [21], [22], [21.5]])
```

```python
# Probabilidades de precisar levar casaco
reg1.predict_proba([[10], [1], [20], [21], [22], [21.5]])
# Colunas: FALSE , TRUE
```

Exemplo de utilização de K-means, com dados gerados automaticamente pelo python, para classificação (clusterização).

```python
# Importando bibliotecas necessárias para trabalhar com os dados e utilizar o modelo de ML, gerar os dados e divisão dos dados
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
```

```python
# Gerando os dados com 300 amostras e 4 pontos centrais (grupos) de distribuição e garantindo repetibilidade
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
# Realizando o plot dos pontos e colorindo conforme o grupo ao qual pertencem
plt.scatter(X[:,0], X[:,1],s=50, c=y_true, cmap='viridis')
```

```python
# Criando modelo de KMeans para classificar em 4 clusters e garantindo repetibilidade
kmeans = KMeans(n_clusters=4, random_state=0)
# Treinando com os dados
kmeans.fit(X)
```

```python
# Realizando a classificação sobre os mesmos dados
y_kmeans = kmeans.predict(X)
```

```python
# Plotando colorindo conforme os grupos classificados pelo K-means
plt.scatter(X[:,0], X[:,1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:,0], centers[:,1], c='black', s=200, alpha=0.5)
# Percebe-se que a classificação está correta e os centros dos clusters foram 
# estimados corretamente no centro dos grupos
```

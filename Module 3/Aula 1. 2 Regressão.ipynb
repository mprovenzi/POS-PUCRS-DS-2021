Exemplo básico utilizando um modelo de Regressão Logística para prever o nível de felicidade de países de acordo com o PIB. 

```python
# Importando bibliotecas necessárias para trabalhar com os dados e utilizar o modelo de ML
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
```

```python
# Importação dos arquivos csv que contém os dados, armazenados no Google Drive

# Importando biblioteca e montando google drive, para ter acesso aos arquivos
from google.colab import drive
drive.mount('/content/gdrive')

# Realizando a leitura dos arquivos
gdrive_path = '/content/gdrive/My Drive/Data Science/Python/2021-22 PUCRS Pós/Disciplina 3 Python para CD/Prof Mauricio SOFTPLAN/Arquivos/'
money_data_frame = pd.read_csv(gdrive_path + 'money.csv')
happiness_data_frame = pd.read_csv(gdrive_path + 'happiness.csv')
```

```python
# Visualização rápida dos dados de GDP (PIB) dos países 
money_data_frame.head(10)
```

```python
# Visualização rápida dos dados de felicidade média dos habitantes dos países
happiness_data_frame.head(10)
# Percebe-se que há países que constam numa lista e não na outra
```

```python
# Criando novo dataframe, a partir de merge interno: apenas países que constam 
# em ambas listas serão incluídos
money_and_happiness_df = money_data_frame.merge(happiness_data_frame, how='inner')
money_and_happiness_df.head(10)
```

```python
# Confirmando realmente que os dados de GDP e happiness estão presentes em todos exemplos
money_data_frame.describe() # count = 105
happiness_data_frame.describe() # count = 40
money_and_happiness_df.describe() # count = 40
```

Agora que os dados estão contidos no dataframe money_and_happiness_df, será iniciada utilização do modelo de Regressão Logística

```python
# criando objetos np pra ajudar na sintaxe
X_array_money = np.array(money_and_happiness_df['GDP']).reshape(-1,1)   # X em maiusculo pois é uma matriz
y_array_happiness = money_and_happiness_df['Happiness']                 # y minúsculo pois é um array
```

```python
# criando a regressao e fazendo o fit (treinamento)
regression = LinearRegression()
regression.fit(X_array_money, y_array_happiness)
```

```python
# Predizendo o nível de felicidade a partir de 3 dados de PIB
regression.predict([[10000], [1000], [50000]]) 
# array([[5.83601431], [5.57145783], [7.01182088]])
```
